# Zaman Serisi Analizi

### Nedir?
    Zaman serisi verilerin zamana göre , eşit aralıklı ve sıralı olarak gözlemlenen verilerdir.
### Analizin amacı nedir?
    Bu tür verileri analiz ederken asıl amacımız, verisetinden çıkarılacak bütün bilgilerin çıkarılmasından sonra geriye kalan beklenmedik durum yada rassal etkiler üzerinden model kurulmasıdır. Rassal etkiler öngörülemeyen etkilerdir fakat diğer bilgi veren etkiler öngörülebilir olduğu için öngörülebilir etkiler dışında kalan kısmı anlamaya çalışıyoruz. Modellemenin asıl amacı ise belli bir hata ile Anakitleyi oluşturan verilerin teorik dağılımını bulabilmektir.
### Varsayım nedir ve zaman serisi analizindeki varsayımlar nelerdir ?
    İstatistikte analizin parametrik olması için ve o analizin yapılabilmesi için belli başlı koşulları sağlaması gerekir.Bu koşullar sağlanmadığında artık analizimiz parametrik olmayan bir analizdir.Parametrik analizler daha güvenilirdir, etkindir  ve özellikle yansızdır(bias yoktur). Yansız olması en az hata ile tahmini üreteceği anlamına gelmez.Parametrik zaman serisi analizindeki varsayımlarımız :
    1-) Seri durağandır.
    2-) Kalıntılar (hatalar) normal dağılmalıdır.(White Noise)
    3-) Kalıntıların ortlaması 0 olmalıdır.
    4-) Kalıntıların varyansı zaman içinde değişmemelidir(zaman içinde varyans homojen olmalıdır).(Homoscedasticity)
    5-) Kalıntılar arası otokorelasyon olmamalıdır.

## Zaman serisi bileşenleri nelerdir ? 
    Zaman serisi Trend , Sezonsallık , Konjoktürel , Rassal etki olamk üzere dört bileşenden oluşmaktadır. Çarpımsal ve Toplamsal olarak iki türlü bileşen etkisi gözlemlenebilir.

### Trend:
    Trend zaman içinde serinin artış veya azalış yönünde genel gidişatıdır.Trend uzun dönemli bir etki sonucu çıkar yani en az 15 yıllık bir dönem olmalıdır.Deterministik ve Stokastik trend olarak iki çeşidi vardır.Kararlı bir şekilde artış ve azalış olması Deterministik trend varlığını gösterir.Deterministik trend doğrusal , parabolik ,üstel olabilir. Stokastik trend ise belli bir kararlılıkta ilerlemeyen ,(genel tabirle alıp başını giden) artış veya azalış trendleridir.Böyle bir trendin varlığı serinin bir stokastik prosese sahip olduğunu gösterir ve öngörülemez.Stokastik trendler literatürde Random Walk olarak ta geçmektedir.

```python
Image(filename='trends.jpg')
```

### Sezonsallık(seasonality):
    Sezonsallık , zaman serisindeki belli aralıklarda meydana gelen , periyodik dalgalanmalardır.Mevsimsellik olarak ta geçmektedir.Mevsimsellik periyodu daha önceden belirlenmesi gerekir. Örneğin bu aydaki satışlar 12 ay önceki satışlar ile benzer ise burada mevsimsellik periyodu 12'dir.Bazen bukadar kolay belirlenemeyebilir ve böyle durumlarda ayrıştırma(decomposing) veya otokorelasyondan yola çıkarak belirlenebilir.

```python
Image(filename='airpassanger.jpg')
```

### Konjoktürel(Cycling):
    Konjoktürel dalgalanmalar , önceden belirlenemeyen , periyodik olmayan dalgalanmalardır.Konjoktürel dalgalanmalar ekonomik ve sektörel verilerde çok fazla görülmketedir.Bir örnek vermek gerekirse ekonomik kriz bir konjoktürel dalgaya sebep olmaktadır. Bir başka örnek ise deprem olması sonucu sebep olacak beklenmedik dalgalanma da konjoktürel dalgalara örnektir.Mevsimsel dalgaların özel bir halidir ve 5 ila 8 yıl kadar etkileri gözlemlenebilir.Ayrıca çoğu kaynakta rassal etki olarak ta gösterilmektedir.

```python
Image(filename='Cycling.jpg')
```

### Rassal etki(Random Effect)(Residual):
    Rassal etki, temel bileşenlerin olmadığı durumda geriye kalan , rassal olarak meydana gelen değerlerdir.Önceden öngörülemez.Seride meydana gelen hatalar üzerinde inceleme yapılması gerekebilir.Tahmin başarısını kontrol ettiğimiz kısımdır.Varsayımlarımızı hatlar üzerinde incelememizin temel sebebi , kurduğumuz modelin seriyi iyi genelleyebilmesini ve modelde seriden kaynaklı bir sapma olup olmadığını tespit edebilmemizden dolayıdır.Biz varsayım olarak white noise olmasını isteriz.White noise, artıkların  0 ortalama ve sabit varyans ile normal dağılmasıdır.

## Otokorelasyon ve kısmi otokorelasyon nedir ?
    Otokorelasyon bir değişkenin şimdiki dönem ile gecikmeli dönemleri arasındaki korelasyondur.Kısmi otokorelasyon ise şimdiki dönem ile gecikmeli dönemler arasındaki bir dönem ile arasındaki korlasyondur. Kısmi otokorelasyonda şimdiki dönem ile istenilen dönem arasındaki diğer gecikmeli değerler dikkate alınmaz.Otokorelasyon bize serinin yapısı hakkında ve kurulacak model hakkında çok önemli bilgiler vermektedir.Feature selection da korelasyon çok önemli bir belirliyicidir.Zaman serisinde de gecikmeli değerlerin her birini bir değişken olarak düşünürsek biz burada değişken seçimini otokorelasyona ve kısmi otokorelasyona göre yapıyoruz.

## Durağanlık(stationary) nedir ?
    Durağanlık serinin ortalamada , varyansta ve otokorelasyonda değişmeyen seviyede sabit olması haldiir.Stokastik proses içermemesi serinin durağan olduğunun göstergesidir.Yani seri stokastik trend te olduğu gibi random walk içermez.Kurulan model parametreleri üzerinden birim kök testi yaparak serinin durağan olup olmadığı test edilebilir.Seri durağan değil ise yapılan tahminler yüksek R-square değeri vereceği için yanıltıcı olacak, sahte regresyon problemine yol açacaktır.Analiz yapılmadan önce kesinlikle serinin durağan olması gerekir. Durağan olmayan seriyi fark alma işlemi ile durağan hale getirebiliriz.

```python
Image(filename='staitonary.png')
```

## Yapısal kırılma(structural break) nedir ?
    Yapısal kırılma , serinin trendinin veya gidişatının bir olay sonucu yön değiştirmesidir.Genel kurulan modellerde çok büyük sapmalara yol açar.Örneğin tedarik zincirindeki bir firma başka bir firma ile anlaşma yaptığında o tarihten itibaren satışların düzeyinin değişmesi yapısal kırılmaya örnektir.Dummy değişkenler modele yapısal kırılma olduğu öğretilebilir.

```python
Image(filename='structural_breaks.png')
```

## Otoregresif(Autoregressive)(AR) model nedir ?
    Otoregresif model serinin gecikmeli dönemleri ile modellenen regresyon denklemidir.Gecikmeli değerleri değişken olarak düşünürsek bu tür gecikmeli regresyon denklemleri otoregresif olarak adlandırılmaktadır.Kısmi otokorelasyon gecikme sayısını belirlemekte fikir vermektedir.Phi parametresinin 1 ve -1 sınırları dışında kalması birimkök olarak adlandırılır.Birim kök içeren seriler durağan değildir.

```python
Image(filename='6photo.jpg')
```

## Hareketli ortalama(moving average)(MA)modeli nedir?
    Model Artıklarının gecikmeli değerlerinden oluşan modele hareketli ortalama modeli denmektedir. Hareketli ortalama denmesinin sebebi , artıkların gecikmeli ortalaması olduğu içindir.MA modeli geçmişte yaşanan şokların etki süresini göstermektedir.Dolayısıyla Otokorelasyon gecikme sayısı belirlemesinde fikir verir.

    Ar ve Ma modellerinin bir arada bulunduğu modeldir.I değeri ise ıntegrated kısmıdır , yani seri durağan değilse kaçıncı dereceden fark alınacağını gösterir.
    SARIMA modeli ARIMA modelinin sezonsal gecikme eklenmiş halidir.
    ARIMAX modeli exogenous kısmı yani başka bağımsız değişkenlerin eklenmiş halidir.
    SARIMAX modeli hem sezonsal hem de bağımsız değişken eklenmiş daha genel bir modeldir.

```python
from tensorflow.keras.layers import Activation , Dense , LSTM ,Dropout
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
from tensorflow.keras import Sequential , activations
from sklearn.preprocessing import MinMaxScaler
```

```python
Istanbul_train = train.Istanbul.values.reshape(-1,1)
Istanbul_test = test.Istanbul.values.reshape(-1,1)
scaler = MinMaxScaler()
scaler.fit(Istanbul_train)
scaled_train = scaler.transform(Istanbul_train)
scaled_test = scaler.transform(Istanbul_test)
```

```python
#generator = TimeseriesGenerator(scaled_train, scaled_train, length=13, batch_size=1)
def create_dataset(dataset, look_back=1):
    X, Y = [], []
    for i in range(len(dataset)-look_back-1):
        a = dataset[i:(i+look_back), 0]
        X.append(a)
        Y.append(dataset[i + look_back, 0])
    return np.array(X), np.array(Y)
    
look_back = 17
X_train, Y_train = create_dataset(scaled_train, look_back)
X_test, Y_test = create_dataset(scaled_test, look_back)

X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
X_test = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))
```

```python
model = Sequential()
model.add(LSTM(120, activation='relu',recurrent_activation="relu", input_shape=(X_train.shape[1], X_train.shape[2]),return_sequences=True))
model.add(Dense(30))
model.add(LSTM(60,return_sequences=True,activation="relu",recurrent_activation="relu"))
model.add(Dense(30))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
history = model.fit(X_train,Y_train,batch_size=32,epochs=40,validation_data=(X_test,Y_test))
```

```python
plt.figure(figsize=(8,4))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Test Loss')
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epochs')
plt.legend(loc='upper right')
plt.show();
```

```python
x = np.concatenate((X_train,X_test),axis=0)
pred = model.predict(x)

pred = pred.reshape(len(pred),1)
pred = scaler.inverse_transform(pred)
pred = pred.reshape(len(pred),)
pred = pd.Series(pred[-len(test):],index=test.index)

pred.plot(label="predict",grid=True,fontsize=12,lw=1,marker=".",figsize=(20,10),colormap="Dark2")
test.Istanbul.plot(grid=True,label="real")
plt.legend()
plt.show();
```

```python
print("LSTM(17),mean squared error:",round(mean_squared_error(pred,test.Istanbul),5))
```
